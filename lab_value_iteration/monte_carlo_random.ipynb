```python
import helperff
```

```python
env_a_random_small = helperff.make_env(is_slippery=False, map_size=5, proba_frozen=0.85, seed=69)
V_a, N_a, returns_a, lengths_a = helperff.run_mc_experiment(
    env_a_random_small,
    episodes=10000,
    alpha=0.1,
    gamma=0.9,
    first_visit=True,
    greedy=False,
    max_steps=100,
)
helperff.plot_experiment_results(env_a_random_small, V_a, N_a, returns_a, lengths_a, gamma=0.9)
```
