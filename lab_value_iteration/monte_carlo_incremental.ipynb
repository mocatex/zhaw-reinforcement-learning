```python
import helperff
```

```python
env_b_greedy_small = helperff.make_env(is_slippery=False, map_size=5, proba_frozen=0.85, seed=69)
V, N, returns, lengths = helperff.run_mc_experiment(
    env_b_greedy_small,
    episodes=1000,
    alpha=0.1,
    gamma=0.9,
    epsilon=0.9,
    epsilon_decay=0.99,
    first_visit=True,
    greedy=True,
    max_steps=200,
)
helperff.plot_experiment_results(env_b_greedy_small, V, N, returns, lengths, gamma=0.9)
```
